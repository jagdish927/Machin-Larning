```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/content/customer_data.csv')
df
```

```python
df.columns
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['AnnualIncome']=df['AnnualIncome'].fillna(df['AnnualIncome'].median())
```

```python
df['NumberOfPurchases']=df['NumberOfPurchases'].fillna(df['NumberOfPurchases'].mean())
```

```python
np.mean(abs(df['AnnualIncome']-df['AnnualIncome'].mean()))
```

```python
df['TimeSpentOnWebsite']=df['TimeSpentOnWebsite'].fillna(df['TimeSpentOnWebsite'].mean())
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.head(2)
```

```python
df['ProductCategory'].value_counts()
```

```python
sns.pairplot(df,hue='PurchaseStatus')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc=LabelEncoder()
```

```python
df['Gender']=enc.fit_transform(df['Gender'])
df['ProductCategory']=enc.fit_transform(df['ProductCategory'])
df['PurchaseStatus']=enc.fit_transform(df['PurchaseStatus'])
```

```python
df.head()
```

```python
df['LoyaltyProgram']=enc.fit_transform(df['LoyaltyProgram'])
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min=MinMaxScaler()
```

```python
scaled_df = pd.DataFrame(min.fit_transform(df), columns=df.columns)
scaled_df
```

```python
x=df.drop('PurchaseStatus',axis=1)
y=df['PurchaseStatus']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
te=DecisionTreeClassifier()
```

```python
te.fit(x_train,y_train)
```

```python
y_pred=te.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,mean_squared_error,confusion_matrix,classification_report,roc_curve
```

```python
print('accuracy_score : ',accuracy_score(y_test,y_pred))
print('mean_squared_error : ',mean_squared_error(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred),annot=True)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred1=rf.predict(x_test)
```

```python
print('accuracy_score : ',accuracy_score(y_test,y_pred1))
print('mean_squared_error : ',mean_squared_error(y_test,y_pred1))
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred1))
```

```python
print(classification_report(y_test,y_pred1))
```

```python
fpr,tpr,threshold=roc_curve(y_test,y_pred)
fpr1,tpr1,threshold=roc_curve(y_test,y_pred1)
```

```python
plt.plot(fpr,tpr,label='te')
plt.plot(fpr1,tpr1,label='rf')
plt.xlabel('fpr')
plt.ylabel('tpr')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import BaggingClassifier
bag_clf = BaggingClassifier((RandomForestClassifier()), bootstrap=True, n_jobs=-1, oob_score=True)
bag_clf.fit(x_train, y_train)
y_pred3 = bag_clf.predict(x_test)
accuracy_score(y_test, y_pred3)
```
